```python
import os
import folium

print(folium.__version__)
```

```python
import numpy as np

data = (np.random.normal(size=(100, 3)) *
        np.array([[1, 1, 1]]) +
        np.array([[37.8, -122, 1]])).tolist()
```

```python
from folium.plugins import HeatMap

m = folium.Map([37.871775, -122.274603], zoom_start=6)

HeatMap(data).add_to(m)
m.save(os.path.join('/Users/ryin/Documents/lbnl/pge-adr/results', 'Heatmap.html'))
```
